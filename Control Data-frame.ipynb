```python
df= pd.read_csv('control_1.csv')
df
```

```python
#starting patient 1 from 12:00am 
my_df= df.drop(df.index[0: 540])

#cutting off dataframe on the 7th day
my_df= my_df.drop(df.index[10620:51612])

#creating an ID tag for patient 1
my_df['Id'] =np.nan
my_df['Id'].fillna('control_1', inplace= True)
my_df
```

```python
def load_hourly_activity(frame):
    my_df['total_hour_activity'] = np.nan
    my_df['timestamp'] = my_df['timestamp'].astype(str)
    
    #
    frame['timestamp'] = pd.to_datetime(frame.timestamp)
    frame['timestamp']= frame['timestamp'].dt.strftime('%H:%M')
    
    #adding the total of each hour and putting the values in a new column
    list = []
    score = 0;
    for i, row in frame.iterrows():
        activity = i
        score = score + row['activity']
        if (i + 1) % 60 == 0:
            list.append(score)
            score = 0
    #only include the hour timestamps        
    goodTimes = frame[frame['timestamp'].str.endswith(":00")]
    frame = goodTimes
    
    frame.reset_index(inplace = True)
    frame['total_hour_activity'] = list
    frame = frame.drop(['index', 'activity'], axis =1)
    
    frame_t = frame.T
    header_row = 0
    frame_t.columns = frame_t.iloc[header_row]
    frame_t = frame_t.drop('timestamp')
    frame_t = frame_t.drop('Id')
    frame_t = frame_t.drop('date')
    frame_t = frame_t.reset_index(drop = True)
    frame_t['Id'] =np.nan
    frame_t['Id'].fillna('control', inplace= True)
    frame =frame_t
    return frame

one_df = load_hourly_activity(my_df)
```
